## Matplotlib and Numpy Examples

Here's an example of how you can work on data science projects with packages like numpy and matplotlib in the lab enviornment:

```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
```

```python
np.random.seed(19680801)

N = 10
data = [np.logspace(0, 1, 100) + np.random.randn(100) + ii for ii in range(N)]
data = np.array(data).T
cmap = plt.cm.coolwarm
rcParams['axes.prop_cycle'] = cycler(color=cmap(np.linspace(0, 1, N)))


from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color=cmap(0.), lw=4),
                Line2D([0], [0], color=cmap(.5), lw=4),
                Line2D([0], [0], color=cmap(1.), lw=4)]

fig, ax = plt.subplots(figsize=(10, 5))
lines = ax.plot(data)
ax.legend(custom_lines, ['Cold', 'Warm', 'Hot'])
```

Or with this example that shows how to plot an equation:

```python
#generate data points
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x) + np.cos(x) - np.tan(x)

# Plotting
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label='sin(x)', color='pink', linestyle='-', linewidth=4)  #plot
plt.title('Plot of Equation: sin(x) + cos(x) - tan(x)')
plt.xlabel('x')  #x-axis label
plt.ylabel('sin(x) + cos(x) - tan(x)')  #y-axis label
plt.legend()  
plt.grid(True)
plt.show()  #Display
```

## Pandas Examples

Here's how you can use the pandas package in the lab environment:

```python
import pandas as pd

data = {
  "plant growth (cm)": [20, 4, 93, 80, 32, 12, 6],
  "time (years)": [1, 2, 8, 2, 4, 1, 2]
}

#load data into a Pandas DataFrame object:
df = pd.DataFrame(data)

print(df)
```

Here's example of using pandas for analysis

```python
#create a data frame with sample data about cats
data = {
    'Name': ['Whiskers', 'Shadow', 'Mittens', 'Paws', 'Tiger', 'Ginger', 'Figaro', 'Luna', 'Garfield'],
    'Age': [2, 5, 3, 4, 1, 3, 5, 3, 4],
    'Cute-ness Rating': [9.5, 8.2, 7.9, 8.8, 9.0, 8.6, 7.5, 8.0, 9.8],
    'Breed': ['Siamese', 'Maine Coon', 'Persian', 'Bengal', 'Sphynx', 'Siamese', 'Bengal', 'Ragdoll', 'Shorthair'],
    'Weight': [4.5, 5.2, 3.9, 4.8, 4.0, 5.3, 5.6, 5.8, 3.5]

}

df = pd.DataFrame(data)

#calculate the average age and average weight
average_age = df['Age'].mean()
average_weight = df['Weight'].mean()

print(f'Average Age: {average_age}')
print(f'Average Weight: {average_weight}')
print(' ')

#find the cutest cat
cutest_cat = df[df['Cute-ness Rating'] == df['Cute-ness Rating'].max()]

print('Finding the cutest Cat:')
print(cutest_cat)
print(' ')

#group by breed and calculate the average weight per breed
average_cuteness_by_breed = df.groupby('Breed')['Cute-ness Rating'].mean()

print('Average Cuteness by Breed:')
print(average_cuteness_by_breed)
print(' ')

#add a new column for weight after a hypothetical growth spurt (20% increase)
df['Weight_after_growth_spurt'] = df['Weight'] * 1.20

print('DataFrame with Weight after Growth Spurt:')
print(df)
```
